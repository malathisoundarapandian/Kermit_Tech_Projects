```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Database configuration
db_config = {
    'host': '10.103.13.225',
    'user': 'root',
    'password': 'StreamTra1ningSets!',
    'database': 'ot'
}

# Connect to the database
conn = mysql.connector.connect(**db_config)

# SQL query for sales summary report
query = """
    SELECT
        region,
        SUM(SALES_AMOUNT) AS total_sales_amount
    FROM sales_analysis
    GROUP BY
    region
"""

# Fetch data using Pandas
sales_data = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Display the generated report data
print(report_data)


# Create a pie chart using Seaborn
plt.figure(figsize=(8, 8))
plt.pie(sales_data['total_sales_amount'], labels=sales_data['region'], autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Amount Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle
plt.tight_layout()


plt.show()
```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Database configuration
db_config = {
    'host': '10.103.13.225',
    'user': 'root',
    'password': 'StreamTra1ningSets!',
    'database': 'ot'
}

# Connect to the database
conn = mysql.connector.connect(**db_config)

# SQL query for sales summary report
query = """
    SELECT
        product_name,
        SUM(rating) AS rating
    FROM sales_analysis
    GROUP BY
    product_name
"""

# Fetch data using Pandas
sales_data = pd.read_sql(query, conn)

# Close the database connection
conn.close()

# Display the generated report data
print(report_data)


# Create a bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='product_name', y='rating', color='yellow', data=sales_data,width=0.2)  # Swap x and y
plt.xlabel('Products')  # Adjust the labels
plt.ylabel('Rating')
plt.title('Product Feedback Rating')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent cropping
plt.show()
```
